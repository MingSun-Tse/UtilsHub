```python
from mpl_toolkits import mplot3d
from matplotlib import pyplot as plt
import numpy as np
import math as m
```

```python
fig = plt.figure()
ax3d = plt.axes(projection='3d')

rots = 2
N = 200
theta = np.linspace(0, 2*m.pi*rots, N+1)[:-1]
x = np.cos(theta)
y = -np.sin(theta)
z = np.sin(theta/rots)
ax3d.scatter3D(x, y, z)
```
